```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
reg = linear_model.LinearRegression()
df = pd.read_csv('data.csv', sep=',', usecols = ['radius_mean', 'perimeter_mean'], index_col=False)
x = df['radius_mean'].values.reshape(-1, 1)
y = df['perimeter_mean'].values.reshape(-1, 1)
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.1)
reg.fit(train_x, train_y)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
predicted_y = reg.predict(test_x)
df_preds = pd.DataFrame({'Actual': test_y.squeeze(), 'Predicted': predicted_y.squeeze()})
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
import numpy as np
import matplotlib.pyplot as plt

r2_score(test_y, predicted_y)
```

```python
plt.scatter(test_x, test_y, color="black")
plt.plot(test_x, predicted_y, color="blue", linewidth=3)
plt.show()
```
